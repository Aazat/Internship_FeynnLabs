```python
i = 0
count = []
ind= []
while(i<kaggle.shape[0]):
    row_count = np.sum(kaggle.iloc[i].isna())
    count.append(row_count)
    if(row_count == 0):
        ind.append(i)
    i += 1
ind
```

```python
len(ind)
```

```python
cleaned = kaggle.iloc[ind]    # only considering cleaned 7 parameters for testing models.
```

```python
cleaned = cleaned.reset_index(drop=True)
cleaned
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result_mul = seasonal_decompose(cleaned.iloc[0, :50], model='multiplicative', period=1 ,extrapolate_trend='freq')
```

```python
plt.rcParams.update({'figure.figsize': (10,10)})
result_mul.plot().suptitle('Multiplicative Decompose', fontsize=22)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(cleaned.iloc[1,:50], order=(1, 0, 0)) 
```

```python
cleaned_tr = cleaned.transpose()
cleaned_tr.columns = ["one","two", "three", "four", "five", "six","seven"]
```

```python
model = ARIMA(cleaned_tr["one"].iloc[:50], order=(1, 0, 0)) 
```

```python
results = model.fit()
```

```python
cleaned_tr['Forecast'] = results.fittedvalues
```

```python
cleaned_tr[["one", "Forecast"]].plot()
```

```python
para = cleaned_tr["one"].iloc[:50]
model_p = ARIMA(para, order=(1, 0, 0))
```

```python
result = model_p.fit()
plt.figure(figsize=(16, 12))
plt.plot(result.fittedvalues)
plt.plot(para)
```
